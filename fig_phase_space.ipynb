```python
from kinetic_measure import *
import plot_property_map as ppm

import pynbody
import matplotlib.pyplot as plt
import numpy as np
import scipy.optimize as so
import scipy.stats as ss
from matplotlib.colors import LogNorm
```

```python
data_file = '/scratch/trq/h1.cosmo50/h1.cosmo50PLK.1536gst1bwK1BH.004096'
param_file = '/scratch/trq/h1.cosmo50/h1.cosmo50PLK.1536gst1bwK1BH.param'
s = pynbody.load(data_file, paramfile = param_file)
s.physical_units()
```

```python
h = s.halos()
h0 = h[0]
pynbody.analysis.halo.center(h0,mode='hyb')
#pynbody.analysis.angmom.sideon(h0, cen=(0,0,0))
```

```python
s_filt = s[pynbody.filt.Sphere('50 kpc')]
```

```python
BCG_star = s_filt.star
```

```python
# phase-space analysis


#pos = np.linalg.norm(BCG_star['pos'], axis = 1)
pos = pynbody.derived.r(BCG_star)
vel_r = dot(BCG_star['pos'],BCG_star['vel'])/pos
vel = np.linalg.norm(BCG_star['vel'], axis = 1)
```

```python
#s_filt.rotate_y(-90)
#s_filt.rotate_x(-90)

plt.figure(figsize = (12,15))


rgb = pynbody.plot.stars.render(s_filt, width = '100 kpc', resolution = 100, dynamic_range = 4,plot = False,ret_im = True)

plt.subplot(321)
ppm.plot(BCG_star['pos'][:,0],BCG_star['pos'][:,1], BCG_star['tform'], npix = 100, vmax = 6.5,vmin = 2, cmap = 'jet', log = False, cbar = True)
plt.contour(rgb[:,:,1], colors = 'k', extent = [-50,50,-50,50])

plt.subplot(322)
ppm.plot(BCG_star['pos'][:,0],BCG_star['pos'][:,1], BCG_star['feh'], vmax = 0.01, vmin = -1,npix = 100, cmap = 'jet', log = False, cbar = True)
plt.contour(rgb[:,:,1], colors = 'k', extent = [-50,50,-50,50])


s_filt.rotate_x(90)

rgb = pynbody.plot.stars.render(s_filt, width = '100 kpc', resolution = 100, dynamic_range = 4,plot = False,ret_im = True)

plt.subplot(323)
ppm.plot(BCG_star['pos'][:,0],BCG_star['pos'][:,1], BCG_star['tform'], npix = 100, vmax = 6.5,vmin = 2, cmap = 'jet', log = False, cbar = True)
plt.contour(rgb[:,:,1], colors = 'k', extent = [-50,50,-50,50])

plt.subplot(324)
ppm.plot(BCG_star['pos'][:,0],BCG_star['pos'][:,1], BCG_star['feh'], vmax = 0.01, vmin = -1,npix = 100, cmap = 'jet', log = False, cbar = True)
plt.contour(rgb[:,:,1], colors = 'k', extent = [-50,50,-50,50])


s_filt.rotate_y(90)

rgb = pynbody.plot.stars.render(s_filt, width = '100 kpc', resolution = 100, dynamic_range = 4,plot = False,ret_im = True)

plt.subplot(325)
ppm.plot(BCG_star['pos'][:,0],BCG_star['pos'][:,1], BCG_star['tform'], npix = 100, vmax = 6.5,vmin = 2, cmap = 'jet', log = False, cbar = True)
plt.contour(rgb[:,:,1], colors = 'k', extent = [-50,50,-50,50])

plt.subplot(326)
ppm.plot(BCG_star['pos'][:,0],BCG_star['pos'][:,1], BCG_star['feh'], vmax = 0.01, vmin = -1,npix = 100, cmap = 'jet', log = False, cbar = True)
plt.contour(rgb[:,:,1], colors = 'k', extent = [-50,50,-50,50])

plt.show()
```

```python
dum = plt.hist2d(pos,vel,bins = 100, cmap = 'jet', norm=LogNorm())
plt.xlabel("r [kpc]")
plt.ylabel("|v$_{3D}$|")
plt.title("2D Histogram")
plt.show()

dum = plt.hist2d(pos,vel_r,bins = 100, cmap = 'jet', norm=LogNorm())
plt.xlabel("r [kpc]")
plt.ylabel("v$_{r}$")
plt.title("2D Histogram")
plt.show()
```
